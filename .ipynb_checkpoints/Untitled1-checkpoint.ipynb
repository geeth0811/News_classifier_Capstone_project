```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("BBC News Train.csv")
```

```python
df.head()
```

```python
df['topic'][0]
```

```python
lst1 = df['topic'].tolist()
```

```python
def extractDigits(lst):
    res = []
    for el in lst:
        sub = el.split(', ')
        res.append(sub)
      
    return(res)
lst2 = extractDigits(lst1)
lst2
```

```python
df1 = pd.DataFrame({'topic':lst2})
df1['topic']
```

```python
df['topic'] = df['topic'].map(df1.set_index('topic')['topic'])
```

```python
topic_dict = df.set_index('topic').to_dict()
df1['topic'] = df1['topic'].apply(lambda x: topic_dict[x])
```

```python
df['topic']=df1['topic']
```
